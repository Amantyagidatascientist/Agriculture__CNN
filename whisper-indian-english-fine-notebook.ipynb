```python
import os
from pydub import AudioSegment

def convert_to_wav(input_file, output_file):
    try:
        audio = AudioSegment.from_file(input_file)
        audio.export(output_file, format="wav")
        print(f" Converted: {os.path.basename(output_file)}")
        return True
    except Exception as e:
        print(f"Error converting {input_file}: {e}")
        return False

audio_folder = "audio"
input_files = [f for f in os.listdir(audio_folder)
              if f.endswith((".mp3", ".webm", ".m4a", ".mp4", ".mkv"))]

if input_files:
    print(f"Converting {len(input_files)} files...")
    for input_file in input_files:
        input_path = os.path.join(audio_folder, input_file)
        base_name = os.path.splitext(input_file)[0]
        output_path = os.path.join(audio_folder, f"{base_name}.wav")

        convert_to_wav(input_path, output_path)

    print("All conversions completed!")
else:
    print("No audio files found in the audio folder")
```

```python
import whisper
import os
from pathlib import Path

def format_time(seconds: float) -> str:
    milliseconds = int((seconds % 1) * 1000)
    seconds = int(seconds)
    minutes, seconds = divmod(seconds, 60)
    hours, minutes = divmod(minutes, 60)
    return f"{hours:02}:{minutes:02}:{seconds:02},{milliseconds:03}"

def transcribe_to_srt(audio_path, srt_path, model):
    try:
        print(f"Transcribing: {os.path.basename(audio_path)}")
        result = model.transcribe(audio_path)

        os.makedirs(os.path.dirname(srt_path), exist_ok=True)

        with open(srt_path, "w", encoding="utf-8") as f:
            for i, segment in enumerate(result["segments"], start=1):
                start = format_time(segment["start"])
                end = format_time(segment["end"])
                text = segment["text"].strip()

                f.write(f"{i}\n")
                f.write(f"{start} --> {end}\n")
                f.write(f"{text}\n\n")

        print(f" SRT saved: {os.path.basename(srt_path)}")
        return True

    except Exception as e:
        print(f" Error transcribing {os.path.basename(audio_path)}: {e}")
        return False

def main():
    print(" Loading Whisper model...")
    model = whisper.load_model("large-v3")  
    print(" Model loaded successfully!\n")

    audio_folder = "audio"
    srt_folder = "srt"

    os.makedirs(srt_folder, exist_ok=True)

    wav_files = [f for f in os.listdir(audio_folder) if f.endswith('.wav')]

    if not wav_files:
        print(" No WAV files found in the audio folder")
        return

    print(f"Found {len(wav_files)} WAV file(s) to transcribe:")
    for i, file in enumerate(wav_files, 1):
        print(f"   {i}. {file}")
    print()

    successful = 0
    total_files = len(wav_files)

    for i, wav_file in enumerate(wav_files, 1):
        print(f"[{i}/{total_files}] Processing: {wav_file}")

        audio_path = os.path.join(audio_folder, wav_file)
        srt_filename = os.path.splitext(wav_file)[0] + ".srt"
        srt_path = os.path.join(srt_folder, srt_filename)

        if transcribe_to_srt(audio_path, srt_path, model):
            successful += 1
        print("-" * 50)

    print(f"\n Transcription completed!")
    print(f"Results: {successful}/{total_files} files processed successfully")

    if successful > 0:
        print(f"SRT files saved in: {srt_folder}/")

if __name__ == "__main__":
    main()
```

```python
import os
import json
import librosa
import soundfile as sf
import pysrt
from pathlib import Path
from tqdm import tqdm
import pandas as pd
import re
import glob
import numpy as np 

def clean_text(text):
    text = text.lower()                               
    text = re.sub(r'[<>{}\[\]|\\`~]', '', text)      
    text = re.sub(r'[^\w\s]', '', text)              
    text = re.sub(r'\s+', ' ', text.strip())  
    return text

def segment_audio_with_srt(audio_path, srt_path, output_dir, min_duration=1.0, max_duration=30.0):
    try:
        # Load audio
        audio, sr = librosa.load(audio_path, sr=16000)  # Whisper uses 16kHz
        # Load SRT file
        subs = pysrt.open(srt_path, encoding='utf-8')
        segments = []
        base_name = os.path.splitext(os.path.basename(audio_path))[0]
        
        valid_segments = 0
        total_audio_length = len(audio) / sr
        
        for i, sub in enumerate(subs):
            start_time = sub.start.to_time()
            end_time = sub.end.to_time()
            start_sec = start_time.hour * 3600 + start_time.minute * 60 + start_time.second + start_time.microsecond / 1e6
            end_sec = end_time.hour * 3600 + end_time.minute * 60 + end_time.second + end_time.microsecond / 1e6
            duration = end_sec - start_sec
            
            if (min_duration <= duration <= max_duration and 
                end_sec <= total_audio_length and 
                start_sec < end_sec):
                
                start_sample = int(start_sec * sr)
                end_sample = int(end_sec * sr)
                
                if end_sample <= len(audio) and start_sample >= 0:
                    segment_audio = audio[start_sample:end_sample]
                    
                    if np.max(np.abs(segment_audio)) > 0.001:
                        segment_filename = f"{base_name}_segment_{i:04d}.wav"
                        segment_path = os.path.join(output_dir, segment_filename)
                        sf.write(segment_path, segment_audio, sr)
                        
                        text = clean_text(sub.text)
                        
                        if text.strip() and len(text.strip()) > 2:  # At least 3 characters
                            segments.append({
                                'audio_path': segment_path,
                                'text': text,
                                'duration': duration
                            })
                            valid_segments += 1
        
        print(f"Created {valid_segments} valid segments from {os.path.basename(audio_path)}")
        return segments
        
    except Exception as e:
        print(f"Error processing {audio_path}: {e}")
        return []

def prepare_training_data():
    training_data = []
    audio_files = glob.glob("./audio/*.wav")
    srt_files = glob.glob("./srt/*.srt")
    
    output_dir = "./segmentedAudio"
    os.makedirs(output_dir, exist_ok=True)
    
    audio_srt_pairs = []
    for audio_file in audio_files:
        audio_name = os.path.splitext(os.path.basename(audio_file))[0]
        matching_srt = None
        for srt_file in srt_files:
            srt_name = os.path.splitext(os.path.basename(srt_file))[0]
            if audio_name == srt_name:
                matching_srt = srt_file
                break
        if matching_srt:
            audio_srt_pairs.append((audio_file, matching_srt))
        else:
            print(f" No matching SRT found for {audio_file}")
    
    
    for audio_path, srt_path in tqdm(audio_srt_pairs, desc="Segmenting audio"):
        segments = segment_audio_with_srt(audio_path, srt_path, output_dir)
        training_data.extend(segments)
    
    if training_data:
        durations = [seg['duration'] for seg in training_data]
        print(f"Average duration: {np.mean(durations):.2f}s")
        print(f"Duration range: {np.min(durations):.2f}s - {np.max(durations):.2f}s")
        
        for i in range(min(3, len(training_data))):
            seg = training_data[i]
            print(f"  {i+1}. Duration: {seg['duration']:.2f}s")
            print(f"     Text: '{seg['text']}'")
            print(f"     Audio: {seg['audio_path']}")
    
    os.makedirs('./trainingdata', exist_ok=True)
    manifest_path = './trainingdata/manifest.json'
    with open(manifest_path, 'w', encoding='utf-8') as f:
        json.dump(training_data, f, indent=2, ensure_ascii=False)
    
    return training_data

training_segments = prepare_training_data()
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import WhisperFeatureExtractor, WhisperTokenizer
import torchaudio
import numpy as np

class WhisperDataset(Dataset):
    def __init__(self, manifest_path, feature_extractor, tokenizer, max_length=448):
        with open(manifest_path, 'r', encoding='utf-8') as f:
            self.data = json.load(f)
        self.feature_extractor = feature_extractor
        self.tokenizer = tokenizer
        self.max_length = max_length
        
        print(f" Dataset loaded: {len(self.data)} samples")
        if len(self.data) == 0:
            raise ValueError("Empty dataset! Check your manifest file.")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        
        audio_path = item['audio_path']
        if not os.path.exists(audio_path):
            raise FileNotFoundError(f"Audio file not found: {audio_path}")
        
        try:
            waveform, sample_rate = torchaudio.load(audio_path)
            
            if waveform.shape[0] > 1:
                waveform = waveform.mean(dim=0, keepdim=True)
            if sample_rate != 16000:
                resampler = torchaudio.transforms.Resample(sample_rate, 16000)
                waveform = resampler(waveform)
            
            waveform_numpy = waveform.squeeze().numpy()
            if len(waveform_numpy) == 0:
                raise ValueError(f"Empty audio file: {audio_path}")
            
            input_features = self.feature_extractor(
                waveform_numpy,
                sampling_rate=16000,
                return_tensors="pt"
            ).input_features.squeeze()
            
            text = item['text']
            if not text.strip():
                raise ValueError(f"Empty text for audio: {audio_path}")
                
            labels = self.tokenizer(
                text,
                max_length=self.max_length,
                padding="max_length",
                truncation=True,
                return_tensors="pt"
            ).input_ids.squeeze()
            
            labels[labels == self.tokenizer.pad_token_id] = -100
            
            return {
                'input_features': input_features,
                'labels': labels
            }
            
        except Exception as e:
            dummy_audio = torch.zeros((80, 3000)) 
            dummy_labels = torch.full((self.max_length,), -100)
            return {
                'input_features': dummy_audio,
                'labels': dummy_labels
            }

def collate_fn(batch):
    """Simple collate function for DataLoader"""
    input_features = torch.stack([item['input_features'] for item in batch])
    labels = torch.stack([item['labels'] for item in batch])
    return {
        'input_features': input_features,
        'labels': labels
    }
```

```python
# ✅ FIXED: Proper imports and setup
import torch
import evaluate
from transformers import (
    WhisperForConditionalGeneration,
    WhisperFeatureExtractor,
    WhisperTokenizer,
    WhisperProcessor,
    Seq2SeqTrainer,
    Seq2SeqTrainingArguments
)
from transformers.models.whisper.english_normalizer import EnglishTextNormalizer
from peft import get_peft_model, LoraConfig
from dataclasses import dataclass
from typing import Any, Dict, List, Union
import os
import numpy as np
import json



model_name = "openai/whisper-tiny"
device = "cuda" if torch.cuda.is_available() else "cpu"
feature_extractor = WhisperFeatureExtractor.from_pretrained(model_name)
tokenizer = WhisperTokenizer.from_pretrained(model_name)
processor = WhisperProcessor.from_pretrained(model_name)

model = WhisperForConditionalGeneration.from_pretrained(
    model_name,
    torch_dtype=torch.float32,  
)

model.config.forced_decoder_ids = None
model.config.suppress_tokens = []
model.config.use_cache = False 
model.config.max_length = 448              
model.config.num_beams = 1                
model.config.do_sample = False            
model.config.temperature = 1.0             
model.config.repetition_penalty = 1.1     

model.config.forced_decoder_ids = processor.get_decoder_prompt_ids(
    language="en",
    task="transcribe"
)

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.1,
    bias="none"
)

model = get_peft_model(model, lora_config)
model.enable_input_require_grads()
model = model.to(device)

def create_proper_splits(manifest_path, train_ratio=0.9):
    with open(manifest_path, 'r') as f:
        data = json.load(f)
    
    if len(data) < 10:
        print("Dataset too small! Need at least 10 samples.")
        return data, data  
    
    # Shuffle data first
    import random
    random.shuffle(data)
    
    split_idx = int(len(data) * train_ratio)
    train_data = data[:split_idx]
    eval_data = data[split_idx:]
    
    if len(eval_data) < 5:
        eval_data = data[-10:]  
        train_data = data[:-10]  
    
    os.makedirs('./trainingdata', exist_ok=True)
    with open('./trainingdata/train_manifest.json', 'w') as f:
        json.dump(train_data, f, indent=2)
    
    with open('./trainingdata/eval_manifest.json', 'w') as f:
        json.dump(eval_data, f, indent=2)
    
    return train_data, eval_data

if len(training_segments) > 0:
    train_data, eval_data = create_proper_splits('./trainingdata/manifest.json')
    train_dataset = WhisperDataset('./trainingdata/train_manifest.json', feature_extractor, tokenizer)
    eval_dataset = WhisperDataset('./trainingdata/eval_manifest.json', feature_extractor, tokenizer)
else:
    exit(1)

print(f"   Training: {len(train_dataset)} samples")
print(f"   Evaluation: {len(eval_dataset)} samples")
```

```python
@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any
    
    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        input_features = [{"input_features": feature["input_features"]} for feature in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")
        
        label_features = [{"input_ids": feature["labels"]} for feature in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")
        
        labels = labels_batch["input_ids"]
        if "attention_mask" in labels_batch:
            labels = labels.masked_fill(labels_batch["attention_mask"].ne(1), -100)
        
        if labels.size(1) > 1: 
            bos_token_id = self.processor.tokenizer.bos_token_id
            if bos_token_id is not None:
                first_tokens = labels[:, 0]
                if torch.all(first_tokens == bos_token_id):
                    labels = labels[:, 1:]
        
        batch["labels"] = labels
        return batch


def compute_metrics(pred):
    
    pred_ids = pred.predictions
    label_ids = pred.label_ids
    
    print(f"Predictions shape: {pred_ids.shape}")
    print(f"Labels shape: {label_ids.shape}")
    
    if len(pred_ids.shape) == 3:
        pred_ids = np.argmax(pred_ids, axis=-1)
    
    label_ids[label_ids == -100] = tokenizer.pad_token_id
    
    try:
        pred_str = tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
        label_str = tokenizer.batch_decode(label_ids, skip_special_tokens=True)
        
        
        for i in range(min(5, len(pred_str))):
            print(f"Pred {i}: '{pred_str[i][:100]}...' (truncated)")
            print(f"Ref  {i}: '{label_str[i]}'")
            print()
        
        normalizer = EnglishTextNormalizer(english_spelling_mapping={})
        
        valid_pairs = []
        hallucination_count = 0
        
        for p, r in zip(pred_str, label_str):
            pred_clean = p.strip()
            ref_clean = r.strip()
            
            if not ref_clean:
                continue
                
            if len(pred_clean) > 100:  
                words = pred_clean.split()
                if len(words) > 10:
                    # Count repeated words
                    word_counts = {}
                    for word in words:
                        word_counts[word] = word_counts.get(word, 0) + 1
                    
                    max_word_freq = max(word_counts.values()) if word_counts else 0
                    if max_word_freq > len(words) * 0.3:
                        hallucination_count += 1
                        pred_clean = ""
            
            valid_pairs.append((pred_clean, ref_clean))
        
        
        if len(valid_pairs) == 0:
            return {"wer": 100.0}
        
        pred_clean, ref_clean = zip(*valid_pairs)
        
        try:
            pred_normalized = []
            ref_normalized = []
            
            for p, r in zip(pred_clean, ref_clean):
                try:
                    pred_normalized.append(normalizer(p).strip())
                    ref_normalized.append(normalizer(r).strip())
                except:
                    pred_normalized.append(p.lower().strip())
                    ref_normalized.append(r.lower().strip())
            
        except Exception as norm_error:
            pred_normalized = [p.lower().strip() for p in pred_clean]
            ref_normalized = [r.lower().strip() for r in ref_clean]
        
        final_pairs = [(p, r) for p, r in zip(pred_normalized, ref_normalized) if r]
        
        if not final_pairs:
            return {"wer": 100.0}
        
        pred_final, ref_final = zip(*final_pairs)
        
        # Calculate WER
        import evaluate
        metric = evaluate.load("wer")
        wer_score = metric.compute(predictions=list(pred_final), references=list(ref_final))
        wer_percentage = wer_score * 100
        
        
        return {"wer": min(wer_percentage, 100.0)}
        
    except Exception as e:
        import traceback
        traceback.print_exc()
        return {"wer": 100.0}

# Initialize components
data_collator = DataCollatorSpeechSeq2SeqWithPadding(processor=processor)
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir="./whisper-fixed",
    
    per_device_train_batch_size=1,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=8,
    
    learning_rate=5e-6, 
    warmup_steps=200,
    max_grad_norm=0.5,
    
    num_train_epochs=1,
    
    eval_strategy="steps",
    eval_steps=25, 
    save_strategy="steps",
    save_steps=50,
    
    load_best_model_at_end=True,
    metric_for_best_model="wer",
    greater_is_better=False, 
    predict_with_generate=True,
    generation_max_length=448,
    fp16=False,  
    dataloader_num_workers=0,
    dataloader_pin_memory=False,
    remove_unused_columns=False,
    logging_steps=5,
    report_to=[],
    local_rank=-1,
)
```

```python
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.train()
```
